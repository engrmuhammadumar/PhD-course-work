## 1. Implement simple linear regression

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
torch.manual_seed(1)
```

```python
# Data
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
```

```python
# Declare and initialize the model: simple linear regression (input_dim=1, output_dim=1).
model = nn.Linear(1,1)
```

```python
print(list(model.parameters()))
```

```python
# Set up the optimizer: use SGD for gradient descent; lr (learning rate) = 0.01
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
# Repeat gradient descent 2,000 times over the full training set
nb_epochs = 2000
for epoch in range(nb_epochs+1):

    # Compute H(x)
    prediction = model(x_train)

    # Compute cost
    cost = F.mse_loss(prediction, y_train) # <== Mean squared error function provided by PyTorch

    # The part that updates H(x) using the cost
    # Initialize gradients to zero
    optimizer.zero_grad()
    # Compute gradients by differentiating the cost function
    cost.backward() # Backward pass
    # Update W and b
    optimizer.step()

    if epoch % 100 == 0:
    # Print logs every 100 steps
      print('Epoch {:4d}/{} Cost: {:.6f}'.format(
          epoch, nb_epochs, cost.item()
      ))
```

```python
# Define an arbitrary input 4
new_var = torch.FloatTensor([[4.0]])
# Get the prediction y for input 4 and store it in pred_y
pred_y = model(new_var) # Forward pass
# Since y = 2x, if the input is 4, y should be close to 8 when training is done correctly
print("Predicted value for input 4 after training:", pred_y)
```

```python
print(list(model.parameters()))
```
