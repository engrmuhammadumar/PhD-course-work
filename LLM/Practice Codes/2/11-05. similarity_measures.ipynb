# 1. Euclidean Distance


```python
import numpy as np
```

```python
def dist(x,y):   
    return np.sqrt(np.sum((x-y)**2))

doc1 = np.array((2,3,0,1))
doc2 = np.array((1,2,3,1))
doc3 = np.array((2,1,2,2))
docQ = np.array((1,1,0,1))

print('Distance between Document 1 and Document Q:',dist(doc1,docQ))
print('Distance between Document 2 and Document Q:',dist(doc2,docQ))
print('Distance between Document 3 and Document Q:',dist(doc3,docQ))
```

# 2. Jaccard Similarity


```python
doc1 = "apple banana everyone like likey watch card holder"
doc2 = "apple banana coupon passport love you"

tokenized_doc1 = doc1.split()
tokenized_doc2 = doc2.split()

print('Document 1:',tokenized_doc1)
print('Document 2:',tokenized_doc2)
```

```python
union = set(tokenized_doc1).union(set(tokenized_doc2))
print('Union of Document 1 and Document 2:',union)
```

```python
intersection = set(tokenized_doc1).intersection(set(tokenized_doc2))
print('Intersection of Document 1 and Document 2:',intersection)
```

```python
print('Jaccard similarity:',len(intersection)/len(union))
```
